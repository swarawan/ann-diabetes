```python
## definisi data train and data test

import pandas
import numpy

data_frame = pandas.read_csv('./diabetes_latih.csv')
data_train_value = data_frame.values
data_train_value = numpy.delete(data_frame, 8, axis=1)
data_train_label = data_frame['Outcome'].values

data_frame = pandas.read_csv('./diabetes_uji.csv')
data_uji_value = data_frame.values
data_uji_value = numpy.delete(data_frame, 8, axis=1)
data_uji_label = data_frame['Outcome'].values

print(data_uji_label)
```

```python
## normalisasi data train and data test

from sklearn.preprocessing import MinMaxScaler

data_train_value = MinMaxScaler().fit_transform(data_train_value)
data_uji_value = MinMaxScaler().fit_transform(data_uji_value)

print(data_uji_value)
```

```python
## klasifikasi menggunakan neural network
## definisi hidden layer, learning rate, epoch (iterasi)

from sklearn.neural_network import MLPClassifier

classifier = MLPClassifier(hidden_layer_sizes=3, learning_rate_init=0.1, max_iter=100)
classifier.fit(data_train_value, data_train_label)

print(classifier)
```

```python
## prediksi

data_uji_predict = classifier.predict(data_uji_value)

print(data_uji_predict)
```

```python
## evaluasi nilai akurasi dr model

from sklearn.metrics import accuracy_score

round(accuracy_score(data_uji_label, data_uji_predict), 3)
```

```python
from sklearn.metrics import RocCurveDisplay
svc_disp = RocCurveDisplay.from_estimator(classifier, data_uji_value, data_uji_label)
```
